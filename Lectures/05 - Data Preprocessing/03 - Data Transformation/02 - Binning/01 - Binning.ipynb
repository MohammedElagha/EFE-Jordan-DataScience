# Binning

**Binning** is process of dividing values of numeric attribute into groups/buckets, to reduce computations and minimize the effects of small observation errors.

**2 Methods of Binning:**
1. Equal Depth/Frequency Binning (Quantile-based discretization) : bins have equal frequency.
2. Equal Width/Distance Binning (Uniform Discretization) : bins have equal width with a range of each bin are defined as [min + w], [min + 2w] …. [min + nw] where w = (max – min) / (no of bins).

##### Equal-Frequence Binning (Auto)

```
df['Customer_Age_b1'] = pd.qcut(df['Customer_Age'], q=4)
```

```
df['Customer_Age_b1']
```

```
df['Customer_Age_b1'].describe()
```

```
df['Customer_Age_b1'].value_counts()
```

Bins Labeling

```
age_b2_labels = ['Maturity', 'Manhood', 'Old', 'Aging']
```

```
df['Customer_Age_b2'] = pd.qcut(df['Customer_Age'], q=4, labels=age_b2_labels)
```

```
df['Customer_Age_b2']
```

```
df['Customer_Age_b2'].describe()
```

```
df['Customer_Age_b2'].value_counts()
```

##### Manual Defined Ranges Binning

```
age_b5_labels = ['Maturity', 'Manhood', 'Old', 'Aging']
```

```
age_bins = [25, 30, 45, 60, 75]
```

```
df['Customer_Age_b5'] = pd.cut(df['Customer_Age'], bins=age_bins, labels=age_b5_labels)
```

```
df['Customer_Age_b5']
```
