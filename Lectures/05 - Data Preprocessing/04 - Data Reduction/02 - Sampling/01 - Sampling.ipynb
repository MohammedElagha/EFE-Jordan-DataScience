### Stratified Sampling

In this type, you must choose base column on which sampling weights are stratified.  
You can choose X or traget column.

```
n = 6
total_count = df['grade'].count()
```

```
(df.groupby('grade').count() / df['grade'].count()) * n
```

```
df.groupby('grade').apply(lambda x:x.sample(n=round((len(x)/total_count)*n)))
```

```
df.groupby('grade').apply(lambda x:x.sample(n=round((len(x)/total_count)*n))).sample(frac=1).reset_index(drop=True)
```
