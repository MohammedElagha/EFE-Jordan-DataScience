```
for i in index:
  for col in columns:
    if type(df.iloc[i][col]) == str:
      df.iloc[i][col] = 0
```

```
for i in index:
  for col in columns:
    if str(df.iloc[i][col]) == 'object':
      df.iloc[i][col] = 0
```
