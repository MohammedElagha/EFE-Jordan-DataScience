### Data Preprocessing

##### Drop Irrelevant Attributes

```
dropped_list = ['school', 'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'nursery', 'G1', 'G2']
```

```
for col in dropped_list:
  df = df.drop(col, axis=1)
```

##### Convert Categorial Attributes for int

```
for col in df.columns:
  if df[col].dtype == 'object':
    df[col] = df[col].astype('category').cat.codes
```

##### Scaling (for only 'absences')

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['absences'] = scaler.fit_transform(df['absences'].to_numpy().reshape(-1, 1))
```

```
df['absences']
```

##### Split into X and Y

```
X = df[['sex', 'address', 'famsize', 'Pstatus', 'studytime', 'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'higher', 
        'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences']]
Y = df['G3']
```



---



### Modeling

##### split into training and test set

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

##### init model

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

##### train model

```
model.fit(x_train, y_train)
```

##### test model

```
y_pred = model.predict(x_test)
```

##### evaluate model

```
from sklearn import metrics
import numpy as np
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
