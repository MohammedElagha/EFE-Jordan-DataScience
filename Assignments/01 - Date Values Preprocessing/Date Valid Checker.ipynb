### Check
date is valid & start_date <= due_date

```
from datetime import datetime
```

```
for index in df.index:
  start_date = df.iloc[index]['start_date']
  due_date = df.iloc[index]['due_date']
  activity = df.iloc[index]['activity']

  print(activity)
  
  try :
    datetime.strptime(start_date, '%d/%b/%Y')
    print(start_date)
  except Exception as e:
    print(start_date + ' - ' + str(e))

  try :
    datetime.strptime(due_date, '%d/%b/%Y')
    print(due_date)
  except Exception as e:
    print(due_date + ' - ' + str(e))

  try :
    if (datetime.strptime(due_date, '%d/%b/%Y') >= datetime.strptime(start_date, '%d/%b/%Y')):
      print('Right Range')
    else:
      print('Wrong Range')
  except Exception as e:
    print('Wrong Date/s')

  print('----------------')
```
